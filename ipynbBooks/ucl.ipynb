```python
cats=['school',
 'sex',
 'address',
 'famsize',
 'Pstatus',
 'Mjob',
 'Fjob',
 'reason',
 'guardian',
 'schoolsup',
 'famsup',
 'paid',
 'activities',
 'nursery',
 'higher',
 'internet',
 'romantic']

pValues=[]
for col in cats:
    groupsizes = df.groupby([col, 'G3']).size()
    ctsum = groupsizes.unstack(col)
    ctsum.fillna(0, inplace=True)
    obs=ctsum.values
    p=scipy.stats.chi2_contingency(obs)[1]
    #pValues.append({col:float(p)*100})
    if p<0.05:
        pValues.append({col:['Dependent', float(p)*100]})
    else:
        pValues.append({col:['Independent', float(p)*100]})
```
